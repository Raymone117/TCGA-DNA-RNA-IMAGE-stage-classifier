```python
import openslide
import numpy as np
from PIL import Image
import os

# 输入路径
svs_path = r"F:\5703Dataset\IMAGE\TCGA_BRCA_images\a740881f-cf92-4577-87b3-893bfe43dd29\TCGA-CG-4460-01A-01-TS1.03d2eb00-e2f3-4c1d-884f-570d961fec15.svs"

# 输出路径
out_dir = r"F:\5703Dataset\patches_TCGA-CG-4460"
os.makedirs(out_dir, exist_ok=True)

# 打开 WSI
slide = openslide.OpenSlide(svs_path)
W, H = slide.dimensions

# 参数
PATCH_SIZE = 256
STEP = 256
WHITE_MEAN_CUTOFF = 240  # 过滤白块

records = []  # 保存 patch 信息和标签

patch_id = 0
for y in range(0, H - PATCH_SIZE + 1, STEP):
    for x in range(0, W - PATCH_SIZE + 1, STEP):
        patch = slide.read_region((x, y), 0, (PATCH_SIZE, PATCH_SIZE)).convert("RGB")
        if np.mean(patch) > WHITE_MEAN_CUTOFF:  # 跳过白块
            continue
        fname = f"patch_{patch_id}_x{x}_y{y}.png"
        fpath = os.path.join(out_dir, fname)
        patch.save(fpath)

        # 每个 patch 的标签固定为 STAGE IV
        records.append([fpath, "STAGE IV"])
        patch_id += 1

print(f"✅ 拆分完成，共生成 {patch_id} 个 patch，全部打标签 STAGE IV")

# 保存 patch 对应的标签清单
import pandas as pd
df = pd.DataFrame(records, columns=["patch_path", "label"])
df.to_csv(os.path.join(out_dir, "patch_manifest.csv"), index=False, encoding="utf-8-sig")
print("标签清单已保存：patch_manifest.csv")
```

```python
import openslide
import numpy as np
from PIL import Image
from skimage import color, filters, morphology
import os
import pandas as pd

# 输入 WSI (Stage III)
svs_path = r"F:\wechat\xwechat_files\wxid_lzg2i457vr3712_f340\msg\file\2025-08\TCGA_BRCA_IMAGE\TCGA_BRCA_IMAGE\b052b2c3-fa18-4706-92e3-8137af39226c\TCGA-EW-A1IZ-01A-01-TSA.a52af132-ed54-40a0-931b-913928cf057f.svs"
out_dir = r"F:\5703Dataset\Stage 3"
os.makedirs(out_dir, exist_ok=True)

slide = openslide.OpenSlide(svs_path)
W, H = slide.dimensions

# ========== Step1: 在低分辨率层生成组织掩膜 ==========
level = 2  # 使用较低分辨率层（更快）
low_img = slide.read_region((0, 0), level, slide.level_dimensions[level]).convert("RGB")
low_np = np.array(low_img)

# 转灰度
gray = color.rgb2gray(low_np)
# Otsu 阈值分割
thresh = filters.threshold_otsu(gray)
mask = gray < thresh  # True=组织，False=背景
# 去噪 + 填洞
mask = morphology.remove_small_objects(mask, 50)
mask = morphology.remove_small_holes(mask, 50)

# ========== Step2: 切 patch 并用掩膜过滤 ==========
PATCH_SIZE = 256
STEP = 256
label = "Stage 3"

downsample = int(slide.level_downsamples[level])  # 对应缩放因子
records = []
patch_id = 0

for y in range(0, H - PATCH_SIZE + 1, STEP):
    for x in range(0, W - PATCH_SIZE + 1, STEP):
        # 计算对应低分辨率坐标
        x_low, y_low = x // downsample, y // downsample
        w_low, h_low = PATCH_SIZE // downsample, PATCH_SIZE // downsample

        # 从掩膜里裁切对应区域
        mask_crop = mask[y_low:y_low+h_low, x_low:x_low+w_low]
        tissue_ratio = mask_crop.mean()

        # 如果组织比例太低，就跳过
        if tissue_ratio < 0.6:
            continue

        # 提取 patch
        patch = slide.read_region((x, y), 0, (PATCH_SIZE, PATCH_SIZE)).convert("RGB")
        fname = f"patch_{patch_id}_x{x}_y{y}_{label.replace(' ','')}.png"
        fpath = os.path.join(out_dir, fname)
        patch.save(fpath)

        records.append([fpath, label, tissue_ratio])
        patch_id += 1

print(f"✅ {svs_path} 切割完成，共生成 {patch_id} 个有效 patch")

# 保存清单
df = pd.DataFrame(records, columns=["patch_path", "label", "tissue_ratio"])
df.to_csv(os.path.join(out_dir, "patch_manifest.csv"), index=False, encoding="utf-8-sig")
print("标签清单已保存：patch_manifest.csv")
```

```python
import os, pandas as pd, numpy as np
from PIL import Image
from collections import Counter
from sklearn.model_selection import GroupShuffleSplit

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models

# ========== 1) 数据路径 ==========
STAGE4_DIR = r"F:\5703Dataset\patches_TCGA-CG-4460（stage4 image）"
STAGE3_DIR = r"F:\5703Dataset\patches_TCGA-CG-4476（stage3 image）"

CSV_STAGE4 = os.path.join(STAGE4_DIR, "patch_manifest.csv")   # 你的csv文件名
CSV_STAGE3 = os.path.join(STAGE3_DIR, "patch_manifest.csv")

CSV_PATHS = [CSV_STAGE4, CSV_STAGE3]

# ========== 2) 读取并合并清单 ==========
dfs = []
for p in CSV_PATHS:
    df = pd.read_csv(p)
    if "patch_path" not in df.columns:
        for c in ["file_path","path","image","img_path"]:
            if c in df.columns:
                df = df.rename(columns={c:"patch_path"})
                break
    if "label" not in df.columns:
        raise ValueError(f"{p} 缺少 label 列")
    dfs.append(df)

all_df = pd.concat(dfs, ignore_index=True)
all_df = all_df.dropna(subset=["patch_path","label"])
all_df["patch_path"] = all_df["patch_path"].astype(str)
all_df = all_df[all_df["patch_path"].apply(os.path.exists)].reset_index(drop=True)

print("总样本数:", len(all_df))
print("类别分布:", Counter(all_df["label"]))

# 标签编码
classes = sorted(all_df["label"].unique())
label2id = {c:i for i,c in enumerate(classes)}
id2label = {i:c for c,i in label2id.items()}
all_df["label_id"] = all_df["label"].map(label2id)

# ========== 3) 数据划分 ==========
groups = all_df["slide"] if "slide" in all_df.columns else all_df["label"]
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, val_idx = next(gss.split(all_df, groups=groups))
train_df = all_df.iloc[train_idx].reset_index(drop=True)
val_df   = all_df.iloc[val_idx].reset_index(drop=True)

print("训练集:", len(train_df), Counter(train_df["label"]))
print("验证集:", len(val_df), Counter(val_df["label"]))

# ========== 4) Dataset ==========
class PatchDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df = df.reset_index(drop=True)
        self.transform = transform
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = Image.open(row["patch_path"]).convert("RGB")
        if self.transform: img = self.transform(img)
        return img, int(row["label_id"])

train_tf = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor()
])
val_tf = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()
])

train_ds = PatchDataset(train_df, transform=train_tf)
val_ds   = PatchDataset(val_df,   transform=val_tf)

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True, num_workers=0)
val_loader   = DataLoader(val_ds,   batch_size=64, shuffle=False, num_workers=0)

# ========== 5) 模型 ==========
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet18(weights=None)  # 如果可以联网换成 weights=models.ResNet18_Weights.DEFAULT
model.fc = nn.Linear(model.fc.in_features, len(classes))
model = model.to(device)

# Loss & Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
epochs = 3  # 先跑几个 epoch 测试

# ========== 6) 训练 & 验证 ==========
def evaluate(model, loader):
    model.eval()
    total, correct, running_loss = 0, 0, 0.0
    with torch.no_grad():
        for imgs, y in loader:
            imgs, y = imgs.to(device), y.to(device)
            out = model(imgs)
            loss = criterion(out, y)
            running_loss += loss.item() * imgs.size(0)
            pred = out.argmax(dim=1)
            correct += (pred == y).sum().item()
            total += imgs.size(0)
    return running_loss / total, correct / total

best_val_acc = 0.0
for ep in range(1, epochs+1):
    model.train()
    running_loss = 0.0
    for imgs, y in train_loader:
        imgs, y = imgs.to(device), y.to(device)
        optimizer.zero_grad()
        out = model(imgs)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)

    train_loss = running_loss / len(train_ds)
    val_loss, val_acc = evaluate(model, val_loader)
    print(f"[Epoch {ep}/{epochs}] train_loss={train_loss:.4f} | val_loss={val_loss:.4f} | val_acc={val_acc:.4f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save({
            "model_state": model.state_dict(),
            "label2id": label2id,
            "id2label": id2label
        }, os.path.join(STAGE4_DIR, "best_resnet18.pt"))
        print(f"  ✅ 新最优，保存模型 best_resnet18.pt (val_acc={val_acc:.4f})")

print("训练完成，类别映射:", label2id)
```
